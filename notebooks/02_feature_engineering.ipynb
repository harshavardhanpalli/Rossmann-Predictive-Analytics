```python
import pandas as pd
```

```python
train = pd.read_csv("../data/raw/train.csv")
store = pd.read_csv("../data/raw/store.csv")

train = train.merge(store, on="Store", how="left")
```

```python
train.head()
```

```python
train["Date"] = pd.to_datetime(train["Date"])

train["Year"] = train["Date"].dt.year
train["Month"] = train["Date"].dt.month
train["Day"] = train["Date"].dt.day

train.head()
```

```python
train.shape
```

```python
train.to_csv("../data/processed/train_features.csv", index=False)
```

```python
train["PromoEffect"] = train["Promo"] * train["Promo2"]
```

```python
train["CompetitionDistance"] = train["CompetitionDistance"].fillna(train["CompetitionDistance"].median())
```

```python
train["IsHoliday"] = train["StateHoliday"].apply(lambda x: 1 if x != "0" else 0)
```

```python
train = train.sort_values(["Store", "Date"])

train["Sales_Lag_1"] = train.groupby("Store")["Sales"].shift(1)
```

```python
train["Sales_Rolling_7"] = train.groupby("Store")["Sales"].transform(lambda x: x.rolling(7).mean())
```

```python
train = train.dropna()
```

```python
train.head()
```

```python
train.to_csv("../data/processed/train_features.csv", index=False)
```
