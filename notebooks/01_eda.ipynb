```python
import pandas as pd
```

```python
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")
store = pd.read_csv("../data/raw/store.csv")
```

```python
train = train.merge(store, on="Store", how="left")
test = test.merge(store, on="Store", how="left")
```

```python
train.describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(train["Sales"], bins=50)
plt.title("Sales Distribution")
plt.show()
```

```python
train["Date"] = pd.to_datetime(train["Date"])

daily_sales = train.groupby("Date")["Sales"].mean()

plt.figure(figsize=(12,5))
plt.plot(daily_sales)
plt.title("Daily Sales Trend")
plt.show()
```

```python
train.groupby("Promo")["Sales"].mean()
```

```python
store_sales = train.groupby("Store")["Sales"].mean().sort_values(ascending=False)

store_sales.head(10)
```

```python
train.groupby("StateHoliday")["Sales"].mean()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(train.corr(numeric_only=True), annot=False)
plt.title("Feature Correlation")
plt.show()
```
